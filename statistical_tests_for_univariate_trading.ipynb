```python
#time series analysis
from scipy import stats
from statsmodels.stats.diagnostic import acorr_ljungbox
import numpy as np
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller,kpss
from scipy.stats import chi2_contingency
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from pmdarima import auto_arima
from arch import arch_model
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from pykalman import KalmanFilter
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
def do_the_adfuller_test(data):
  result = adfuller(data['Close'])
  print('ADF test')
  print(f"ADF Statistic:{result[0]}")
  print(f"p-value:{result[1]}")
  print(f"Critical Values:")
  print(f"Lags Used: {result[2]}")
  print(f"Observations: {result[3]}")
  for key,value in result[4].items():
    print(f"{key}:{value}")
  return result[1]
def do_the_kpss_test(data):
  result = kpss(data['Close'])
  print('KPSS test')
  print(f"KPSS Statistic:{result[0]}")
  print(f"p-value:{result[1]}")
  print(f"Critical Values:")
  print(f"Lags Used: {result[2]}")
  print(f"Observations: {result[3]}")
  for key,value in result[3].items():
    print(f"{key}:{value}")
  return result[1]

def parsimony_enhanced(data, test_size=0.2, ljung_lags=[5, 10, 15, 20]):
    """
    Enhanced ARIMA modeling with multiple criteria, out-of-sample validation,
    and comprehensive Ljung-Box testing.

    Parameters:
    -----------
    data : pandas.DataFrame
        DataFrame containing 'Close_log' column
    test_size : float
        Proportion of data to use for out-of-sample testing (0-1)
    ljung_lags : list
        Lags to test for Ljung-Box statistic

    Returns:
    --------
    dict: Comprehensive model results with diagnostics
    """

    # 1. Split data for out-of-sample validation
    split_idx = int(len(data) * (1 - test_size))
    train = data.iloc[:split_idx].copy()
    test = data.iloc[split_idx:].copy()

    print("=" * 70)
    print("ARIMA MODEL SELECTION WITH MULTI-CRITERIA EVALUATION")
    print(f"Training samples: {len(train)}, Test samples: {len(test)}")
    print("=" * 70)

    # 2. Visual diagnostics on training data
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    plot_acf(train['Close_log'], lags=50, ax=axes[0])
    axes[0].set_title(f"ACF (Training, n={len(train)})")
    plot_pacf(train['Close_log'], lags=50, ax=axes[1])
    axes[1].set_title(f"PACF (Training, n={len(train)})")
    plt.tight_layout()
    plt.show()

    # 3. Model selection with different criteria
    criteria = ['aic', 'bic', 'aicc']
    all_results = {}

    for criterion in criteria:
        print(f"\n{'='*60}")
        print(f"OPTIMIZING FOR {criterion.upper()}")
        print('='*60)

        # Auto ARIMA model selection
        model = auto_arima(
            train['Close_log'],
            start_p=0, max_p=5,
            start_q=0, max_q=5,
            d=None,
            seasonal=False,
            stepwise=True,
            suppress_warnings=True,
            trace=True,
            information_criterion=criterion,
            n_jobs=-1
        )

        p, d, q = model.order

        # Fit model
        fitted = SARIMAX(
            train['Close_log'],
            order=(p, d, q)
        ).fit(disp=False)

        # Get residuals
        residuals = fitted.resid.dropna()

        # In-sample diagnostics
        sse = (residuals**2).sum()
        mse = sse / len(residuals)
        rmse_in = np.sqrt(mse)

        # Ljung-Box test on residuals (check for autocorrelation)
        ljung_results = {}
        for lag in ljung_lags:
            lb_test = acorr_ljungbox(residuals, lags=[lag], return_df=True)
            ljung_results[f'lb_lag{lag}'] = {
                'test_stat': lb_test['lb_stat'].iloc[0],
                'p_value': lb_test['lb_pvalue'].iloc[0]
            }

        # Count significant Ljung-Box tests (p < 0.05)
        significant_lb = sum(1 for lag in ljung_lags
                           if ljung_results[f'lb_lag{lag}']['p_value'] < 0.05)

        # Out-of-sample forecast
        forecast = fitted.forecast(steps=len(test))
        forecast_errors = test['Close_log'].values - forecast.values
        rmse_oos = np.sqrt((forecast_errors**2).mean())

        # Store results
        all_results[criterion] = {
            'order': model.order,
            'p': p, 'd': d, 'q': q,
            'params': fitted.params.to_dict(),
            'in_sample': {
                'aic': fitted.aic,
                'bic': fitted.bic,
                'aicc': fitted.aic + (2*len(fitted.params)*(len(fitted.params)+1))/(len(train)-len(fitted.params)-1),
                'sse': sse,
                'rmse': rmse_in,
                'log_likelihood': fitted.llf,
                'converged': fitted.mle_retvals['converged']
            },
            'out_of_sample': {
                'rmse': rmse_oos,
                'mae': np.abs(forecast_errors).mean(),
                'mape': np.mean(np.abs(forecast_errors / test['Close_log'].values)) * 100
            },
            'residuals_diagnostics': {
                'mean': residuals.mean(),
                'std': residuals.std(),
                'skew': residuals.skew(),
                'kurtosis': residuals.kurtosis(),
                'ljung_box': ljung_results,
                'significant_lb_count': significant_lb,
                'lb_lags_tested': ljung_lags
            },
            'model': fitted,
            'forecast': forecast
        }

        # Print summary for this criterion
        print(f"\nARIMA{p,d,q} selected by {criterion.upper()}")
        print(f"In-sample RMSE: {rmse_in:.6f}")
        print(f"Out-of-sample RMSE: {rmse_oos:.6f}")
        print(f"Ljung-Box significant at {significant_lb}/{len(ljung_lags)} lags")
        print(f"AIC: {fitted.aic:.2f}, BIC: {fitted.bic:.2f}")

    # 4. Compare all criteria
    print("\n" + "="*70)
    print("COMPREHENSIVE MODEL COMPARISON")
    print("="*70)

    comparison_data = []
    for criterion in criteria:
        res = all_results[criterion]
        comparison_data.append({
            'Criterion': criterion.upper(),
            'Order': f"({res['p']},{res['d']},{res['q']})",
            'In-sample AIC': f"{res['in_sample']['aic']:.2f}",
            'In-sample BIC': f"{res['in_sample']['bic']:.2f}",
            'In-sample RMSE': f"{res['in_sample']['rmse']:.6f}",
            'Out-sample RMSE': f"{res['out_of_sample']['rmse']:.6f}",
            'Ljung-Box Sig': f"{res['residuals_diagnostics']['significant_lb_count']}/{len(ljung_lags)}",
            'Residual Mean': f"{res['residuals_diagnostics']['mean']:.6f}",
            'Residual Std': f"{res['residuals_diagnostics']['std']:.6f}"
        })

    comparison_df = pd.DataFrame(comparison_data)
    print(comparison_df.to_string(index=False))

    # 5. Select best model based on combined criteria
    print("\n" + "="*70)
    print("MODEL SELECTION RECOMMENDATIONS")
    print("="*70)

    # Rule-based selection
    recommendations = []

    # Check Ljung-Box results (want p > 0.05 for white noise residuals)
    for criterion in criteria:
        res = all_results[criterion]
        sig_count = res['residuals_diagnostics']['significant_lb_count']
        if sig_count == 0:
            recommendations.append(f" {criterion.upper()} model has clean residuals (0/{len(ljung_lags)} significant LB tests)")
        else:
            recommendations.append(f" {criterion.upper()} model has {sig_count}/{len(ljung_lags)} significant LB tests")

    # Find best by out-of-sample RMSE
    best_oos = min(criteria, key=lambda x: all_results[x]['out_of_sample']['rmse'])

    # Find best by AIC
    best_aic = min(criteria, key=lambda x: all_results[x]['in_sample']['aic'])

    # Find best by BIC
    best_bic = min(criteria, key=lambda x: all_results[x]['in_sample']['bic'])

    # Find model with cleanest residuals
    cleanest_resid = min(criteria,
                        key=lambda x: all_results[x]['residuals_diagnostics']['significant_lb_count'])

    print("\nPerformance Summary:")
    print(f"- Best out-of-sample RMSE: {best_oos.upper()} model")
    print(f"- Best AIC: {best_aic.upper()} model")
    print(f"- Best BIC: {best_bic.upper()} model")
    print(f"- Cleanest residuals: {cleanest_resid.upper()} model")

    for rec in recommendations:
        print(f"- {rec}")

    # Final recommendation
    print("\n" + "="*70)
    print("FINAL RECOMMENDATION")
    print("="*70)

    # Decision logic
    if best_oos == best_aic and all_results[best_oos]['residuals_diagnostics']['significant_lb_count'] == 0:
        print(f" Strong recommendation: Use {best_oos.upper()} model")
        print(f"  ARIMA{all_results[best_oos]['order']}")
        print(f"  Reasons: Best OOS performance + Best AIC + Clean residuals")
        final_choice = best_oos
    elif all_results[cleanest_resid]['residuals_diagnostics']['significant_lb_count'] == 0:
        print(f" Recommendation: Use {cleanest_resid.upper()} model for reliable residuals")
        print(f"  ARIMA{all_results[cleanest_resid]['order']}")
        print(f"  Note: Residual whiteness is crucial for valid inference")
        final_choice = cleanest_resid
    else:
        print(f" Trade-off recommendation: Use {best_oos.upper()} model for forecasting")
        print(f"  ARIMA{all_results[best_oos]['order']}")
        print(f"  Warning: {all_results[best_oos]['residuals_diagnostics']['significant_lb_count']}/{len(ljung_lags)} LB tests significant")
        final_choice = best_oos

    # 6. Detailed diagnostics for chosen model
    print("\n" + "="*70)
    print(f"DETAILED DIAGNOSTICS FOR {final_choice.upper()} MODEL")
    print("="*70)

    final_results = all_results[final_choice]
    residuals = final_results['model'].resid.dropna()

    # Detailed Ljung-Box table
    print("\nLjung-Box Test Results (H0: No autocorrelation):")
    print("-" * 60)
    print(f"{'Lag':<8} {'Test Stat':<12} {'p-value':<12} {'Significant':<12}")
    print("-" * 60)

    for lag in ljung_lags:
        lb_info = final_results['residuals_diagnostics']['ljung_box'][f'lb_lag{lag}']
        p_val = lb_info['p_value']
        is_sig = "Yes (p<0.05)" if p_val < 0.05 else "No"
        print(f"{lag:<8} {lb_info['test_stat']:<12.4f} {p_val:<12.4f} {is_sig:<12}")

    print("-" * 60)
    print(f"Significant at {final_results['residuals_diagnostics']['significant_lb_count']}/{len(ljung_lags)} lags")

    # Critical values interpretation
    sig_count = final_results['residuals_diagnostics']['significant_lb_count']
    if sig_count == 0:
        print(" Residuals appear to be white noise (good)")
    elif sig_count <= len(ljung_lags) // 3:
        print(" Mild residual autocorrelation present")
    else:
        print(" Significant residual autocorrelation - model may be misspecified")

    # 7. Visual diagnostics for chosen model
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    fig.suptitle(f'ARIMA{final_results["order"]} - {final_choice.upper()} Model Diagnostics', fontsize=14)

    # Residuals plot
    axes[0, 0].plot(residuals)
    axes[0, 0].axhline(y=0, color='r', linestyle='--', alpha=0.5)
    axes[0, 0].set_title('Residuals over Time')
    axes[0, 0].set_xlabel('Time')
    axes[0, 0].set_ylabel('Residuals')

    # Histogram
    axes[0, 1].hist(residuals, bins=30, edgecolor='black', alpha=0.7, density=True)
    axes[0, 1].set_title('Residual Distribution')
    axes[0, 1].set_xlabel('Residuals')

    # Q-Q plot
    sm.qqplot(residuals, line='45', fit=True, ax=axes[0, 2])
    axes[0, 2].set_title('Q-Q Plot')

    # ACF of residuals
    plot_acf(residuals, lags=30, ax=axes[1, 0])
    axes[1, 0].set_title('ACF of Residuals')

    # PACF of residuals
    plot_pacf(residuals, lags=30, ax=axes[1, 1])
    axes[1, 1].set_title('PACF of Residuals')

    # Ljung-Box p-values across lags
    lags_range = range(1, max(ljung_lags) + 1)
    p_values = [final_results['residuals_diagnostics']['ljung_box'][f'lb_lag{lag}']['p_value']
                if lag in ljung_lags else np.nan
                for lag in lags_range]

    axes[1, 2].plot(lags_range, p_values, 'bo-', label='p-value')
    axes[1, 2].axhline(y=0.05, color='r', linestyle='--', label='0.05 significance')
    axes[1, 2].set_title('Ljung-Box Test p-values')
    axes[1, 2].set_xlabel('Lag')
    axes[1, 2].set_ylabel('p-value')
    axes[1, 2].legend()
    axes[1, 2].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # 8. Return comprehensive results
    return {
        'all_models': all_results,
        'comparison_table': comparison_df,
        'selected_model': final_choice,
        'selected_results': final_results,
        'recommendation': final_choice
    }
def kalman_filtering(data):
    deal = data['Close'].to_numpy()
    kf = KalmanFilter(transition_matrices = [1],
                      observation_matrices = [1],
                      initial_state_mean=deal[0],
                      n_dim_obs=1
                    )
    kf = kf.em(deal,n_iter=10)
    smoothed_state_means, deals = kf.smooth(deal)
    return pd.Series(smoothed_state_means.flatten(),index=data.index)
def holt_winter_smoothing(data):
  # Holt-Winters Exponential Smoothing
  model = ExponentialSmoothing(data['Close'],
                             trend="add",
                             seasonal="add",
                             seasonal_periods=5)
  fitted = model.fit()
  forecast = fitted.forecast(steps=50)  # next 50 days
  print(fitted.fittedvalues)
  plt.figure(figsize=(10,5))
  plt.plot(data['Close'], label="Actual")
  plt.plot(fitted.fittedvalues, label="Fitted")
  plt.plot(forecast, label="Forecast", linestyle="--")
  plt.legend()
  plt.show()
  return forecast
```

```python

#univariate feature engineering
def compute_macd(data, fast=12, slow=26, signal=9):
    prices = data['Close']
    ema_fast = prices.ewm(span=fast, adjust=False).mean()
    ema_slow = prices.ewm(span=slow, adjust=False).mean()
    macd_line = ema_fast - ema_slow
    signal_line = macd_line.ewm(span=signal, adjust=False).mean()
    histogram = macd_line - signal_line
    data["MACD"] = macd_line
    data["MACD_signal"] = signal_line
    data["MACD_hist"] = histogram
    return data
def compute_atr(data, period=14):
    high = data['High']
    low = data['Low']
    close = data['Close']
    high_low = high - low
    high_close = (high - close.shift()).abs()
    low_close = (low - close.shift()).abs()
    tr = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
    atr = tr.rolling(window=period).mean()
    data["ATR"] = atr
    return data
def compute_rsi(data, period=14):
    prices = data['Close']
    delta = prices.diff()
    gain = delta.where(delta > 0, 0.0)
    loss = -delta.where(delta < 0, 0.0)

    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    data['RSI'] = rsi
    return data
def compute_bollinger_bands(data):
  length = 20
  std= 2
  prices = data['Close']
  middle = prices.rolling(window=length).mean()
  rolling_std = prices.rolling(window=length).std()
  upper = middle + std * rolling_std
  lower = middle - std * rolling_std
  data['BB_Upper'] = upper
  data['BB_Lower'] = lower
  data['BB_Middle'] = middle
  return data
def obv_manual(df):
    obv = [0]  # start OBV at 0
    for i in range(1, len(df)):
        # Explicitly get scalar values using .item() to avoid ValueError
        if df['Close'].iloc[i].item() > df['Close'].iloc[i-1].item():
            obv.append(obv[-1] + df['Volume'].iloc[i].item())
        elif df['Close'].iloc[i].item() < df['Close'].iloc[i-1].item():
            obv.append(obv[-1] - df['Volume'].iloc[i].item())
        else:
            obv.append(obv[-1])
    df['OBV'] = obv
    return df
def technical_indicators(data):
  data['SMA_10'] = data['Close'].rolling(window=10).mean()
  data['EMA_10'] = data['Close'].ewm(span=10, adjust=False).mean()
  # 6. Returns
  data['Return'] = data['Close'].pct_change()  # Simple return
  data['LogReturn'] = (data['Close'] / data['Close'].shift(1)).apply(lambda x: np.log(x))
  # 7. Volume Indicators
  data['Volume_MA10'] = data['Volume'].rolling(window=10).mean()
  for lag in [1, 3, 5]:
    data[f'Close_lag{lag}'] = data['Close'].shift(lag)
    data[f'Return_lag{lag}'] = data['Return'].shift(lag)
  data['RollingVol_10'] = data['Return'].rolling(window=10).std()
  data['RollingMax_10'] = data['High'].rolling(window=10).max()
  data['RollingMin_10'] = data['Low'].rolling(window=10).min()
  data.dropna(inplace=True)
  return data
def calculate_hurst_exponent(data,min_window = 2):
  ts = data['Close']
  ts = ts.values.flatten()
  N = len(ts)
  window_sizes = range(min_window, N//2)
  RS_values = []
  window = []
  for n in window_sizes:
    num_subperiods = N // n
    RS_list = []
    for i in range(num_subperiods):
      sub_data = ts[i*n:(i+1)*n]
      # Step 1: Calculate mean
      mean = np.mean(sub_data)
      # Step 2: Mean-adjusted series
      Y = sub_data - mean
      # Step 3: Cumulative deviation
      Z = np.cumsum(Y)
      # Step 4: Range R
      R = np.max(Z) - np.min(Z)
      # Step 5: Standard deviation S
      S = np.std(sub_data, ddof=1)
      # Avoid division by zero
      if S != 0:
          RS_list.append(R / S)
      # Average R/S for this window size
    if RS_list:
        RS_values.append(np.mean(RS_list))
        window.append(n)
  print(window)
  print(RS_values)
  log_n = np.log(window)
  log_RS = np.log(RS_values)
  # Hurst exponent is the slope of log(R/S) vs log(n)
  H = np.polyfit(log_n, log_RS, 1)[0]
  return H
def bullbear(data):
  des = []
  desi=[]
  std_diff = data['Close'].iloc[:, 0]-data['EMA_10']
  tolema = std_diff.std()
  std_diff = data['MACD'] - data['MACD_signal']
  tolmacd = std_diff.std()
  for i in range(len(data)):
    count=0;

    #check with ema

    if data['Close'].iloc[i].item() > data['EMA_10'].iloc[i].item() * (1+tolema):
      count+=1
    elif data['Close'].iloc[i].item() < data['EMA_10'].iloc[i].item() * (1-tolema):
      count-=1
    else :
      count += 0

    #check with rsa
    if data['RSI'].iloc[i].item() < 40:
      count+=1
    elif data['RSI'].iloc[i].item() > 60:
      count-=1
    else:
      count += 0
    # check with obv
    if(i+1<len(data)):
      if data['OBV'].iloc[i].item() < data['OBV'].iloc[i+1].item():
        count+=1
      elif data['OBV'].iloc[i].item() > data['OBV'].iloc[i+1].item():
        count-=1
      else :
        count += 0
    else:
      count+=0

    # check with MACD

    if data['MACD'].iloc[i].item() > data['MACD_signal'].iloc[i].item()*(1+tolmacd):
      count+=1
    elif data['MACD'].iloc[i].item() < data['MACD_signal'].iloc[i].item()*(1-tolmacd):
      count -= 1
    else :
      count += 0

    #final conclusion
    if count>0:
      des.append('Bullish')
      desi.append('Buy')
    elif count<0:
      des.append('Bearish')
      desi.append('Sell')
    else:
      des.append('Neutral')
      desi.append('Hold')
  data['regime_nature'] = des
  data['status'] = desi
  return data
```

```python
def time_series_analysis(data):
  data.dropna()
  price = data['Close']
  data = compute_macd(data)
  data = compute_atr(data)
  data = compute_rsi(data)
  data = compute_bollinger_bands(data)
  data = obv_manual(data)
  data = technical_indicators(data)
  data = bullbear(data)
  # Pass only the 'Close' price series to calculate_hurst_exponent
  hurst_val = calculate_hurst_exponent(data)

  print(hurst_val)
  print(f"Hurst Exponent: {hurst_val}")

  holt_result = holt_winter_smoothing(data)
  print("Holt-Winters forecast generated")

  kalman_result = kalman_filtering(data)
  print("Kalman filtering completed")
  adf_p = do_the_adfuller_test(data)
  kpss_p = do_the_kpss_test(data)
  if adf_p < 0.05 and kpss_p >= 0.05:
    print("Data is Stationary")
  else :
    print("Data is Non-Stationary")
  data['Close_log'] = np.log(data['Close'])
  arima_results = parsimony_enhanced(data, 0.2, [5, 10, 15, 20])
  print(arima_results)

  data["Close"] = price
  return data
```
